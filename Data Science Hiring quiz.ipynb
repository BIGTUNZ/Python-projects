```python
import random
import threading
import time
import os

# List of data science questions (same as before)
questions = [
    {
        "question": "What does 'p-value' in statistical hypothesis testing represent?",
        "options": ["A) Probability of the null hypothesis being true", "B) Probability of obtaining the observed data if the null hypothesis is true", "C) Probability of making a Type I error", "D) All of the above"],
        "answer": "B"
    },
    {
        "question": "Which of the following is not a type of machine learning?",
        "options": ["A) Supervised Learning", "B) Unsupervised Learning", "C) Reinforcement Learning", "D) Deviation Learning"],
        "answer": "D"
    },
    {
        "question": "What is the main purpose of feature engineering?",
        "options": ["A) To increase the size of the dataset", "B) To improve the model's performance", "C) To visualize data", "D) To select the model type"],
        "answer": "B"
    },
    {
        "question": "In regression analysis, what does R-squared represent?",
        "options": ["A) The correlation between variables", "B) The proportion of variance explained by the model", "C) The residual sum of squares", "D) The slope of the regression line"],
        "answer": "B"
    },
    {
        "question": "Which library is primarily used for data manipulation and analysis in Python?",
        "options": ["A) NumPy", "B) Matplotlib", "C) Pandas", "D) Scikit-learn"],
        "answer": "C"
    }
]

def load_high_score():
    if os.path.exists("high_score.txt"):
        with open("high_score.txt", "r") as file:
            return int(file.read().strip())
    return 0

def save_high_score(score):
    with open("high_score.txt", "w") as file:
        file.write(str(score))

def run_quiz(questions):
    score = 0
    random.shuffle(questions) 
    high_score = load_high_score()  

    print(f"Current high score: {high_score}")

    def timer(seconds):
        for i in range(seconds, 0, -1):
            print(f"\rTime left: {i} seconds", end="")
            time.sleep(1)
        print("\nTime's up!")

    for question in questions:
        print("\n" + question["question"])
        for option in question["options"]:
            print(option)

        
        timer_thread = threading.Thread(target=timer, args=(10,))  
        timer_thread.start()
        
        answer = input("\nPlease enter your answer (A, B, C, or D): ").upper()

        if timer_thread.is_alive():
            timer_thread.join(timeout=0)  # Wait for the thread to finish
            if answer in ["A", "B", "C", "D"]:
                if answer == question["answer"]:
                    print("Correct!\n")
                    score += 1
                else:
                    print(f"Incorrect! The correct answer was {question['answer']}.\n")
            else:
                print("Invalid input. Please enter A, B, C, or D.\n")
        else:
            print("You didn't answer in time! The correct answer was:", question["answer"])

    print(f"Your final score is: {score}/{len(questions)}")

    if score > high_score:
        print("Congratulations! You've set a new high score!")
        save_high_score(score)

if __name__ == "__main__":
    print("Welcome to the Data Science Quiz Game!\n")
    run_quiz(questions)
```
